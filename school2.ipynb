import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Données d'entraînement
X = np.array([[100], [110], [120], [130], [140], [150], [160], [170], [180], [190]])
y = np.array([45, 51, 54, 61, 66, 70, 74, 78, 85, 89])

# Affichage de la droite de régression
plt.scatter(y, X)
plt.plot(y, X)
plt.xlabel('Rendement en %')
plt.ylabel('Température en °C')
plt.show()
# Entraînement du modèle de régression linéaire
reg = LinearRegression().fit(X, y)

# Affichage de la droite de régression
plt.scatter(y, X)
plt.plot(y, reg.predict(y.reshape(-1, 1)), color='red')
plt.xlabel('Rendement en %')
plt.ylabel('Température en °C')
plt.show()

# Affichage des coefficients de la droite de régression
print("Coefficient : ", reg.coef_[0])
print("Ordonnée à l'origine : ", reg.intercept_)

# Calcul du coefficient de détermination
R2 = reg.score(X, y)
print("Coefficient de détermination : ", R2)


```python

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Données d'entraînement
x = np.array([[100], [110], [120], [130], [140], [150], [160], [170], [180], [190]])
y = np.array([45, 51, 54, 61, 66, 70, 74, 78, 85, 89])

# Affichage de la droite de régression
plt.scatter(x, y)
plt.plot(x, y)
plt.xlabel('Rendement en %')
plt.ylabel('Température en °C')
plt.show()
# Entraînement du modèle de régression linéaire
reg = LinearRegression().fit(X, y)

# Affichage de la droite de régression
plt.scatter(x, y)
plt.plot(x, reg.predict(y.reshape(-1, 1)), color='red')
plt.xlabel('Rendement en %')
plt.ylabel('Température en °C')
plt.show()

# Affichage des coefficients de la droite de régression
print("Coefficient : ", reg.coef_[0])
print("Ordonnée à l'origine : ", reg.intercept_)

# Calcul du coefficient de détermination
R2 = reg.score(X, y)
print("Coefficient de détermination : ", R2)
```
